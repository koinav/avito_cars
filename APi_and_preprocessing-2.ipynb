```python
import pandas as pd
import sys
import logging
import requests
from time import sleep
from tqdm import tqdm


dataframes = []

for i in range(5, 215, 5):
    file_name = f"cars_data_part_{i}.csv"
    df = pd.read_csv(file_name)
    dataframes.append(df)

df = pd.concat(dataframes, ignore_index=True)

df.to_csv("cars_data.csv", index=False)
```

```python
df.head()
```

```python
columns_to_drop = ["URL", "Руль", "История пробега", "Комплектация", "ПТС", "VIN или номер кузова", "Обмен"]
df = df.drop(columns=columns_to_drop)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df
```

```python
# Изучим категориальные колонки
columns_to_inspect = ["Владельцев по ПТС", "Тип двигателя", "Коробка передач", "Привод", "Тип кузова"]

for col in columns_to_inspect:
    unique_vals = df[col].unique()
    print(f"Уникальные значения для {col}: {unique_vals}")
```

```python
# При выгрузке у нас съехали данные о бренде и модели в некоторых строках. Будем это исправлять
mask = df["Brand"] == "С пробегом"
df.loc[mask, "Brand"] = df.loc[mask, "Model"]
df.loc[mask, "Model"] = df.loc[mask, "Version"]

df.drop("Version", axis=1, inplace=True)

df.head(3)
```

```python
df['Price'] = df['Price'].str.replace('[^\d]', '', regex=True).astype(int)
df['Пробег'] = df['Пробег'].str.replace('[^\d]', '', regex=True).astype(int)
df['Год выпуска'] = df['Год выпуска'].astype(int)
df['Объём двигателя'] = df['Объём двигателя'].str.replace('л', '').str.replace(',', '.').astype(float)
```

```python
# Разделим поле поколение на 3 поля (название поколения, и с какого по какой год выпускалось)
generation_split = df['Поколение'].str.extract(r'^(.*?)\s*\((\d+)[^—]*—(\d+)\)')

df['Generation'] = generation_split[0]
df['Gen_Start'] = generation_split[1].astype(float).astype('Int64')
df['Gen_End'] = generation_split[2].astype(float).astype('Int64')

df = df.drop('Поколение', axis=1)
```

```python
# Извлечем лошадиные силы из поля Модификация
df['Horsepower'] = df['Модификация'].str.extract(r'(\d+)\s*л\.с\.').astype(float)

# Удаляем мощность из оригинальной колонки
df['Модификация'] = df['Модификация'].str.replace(r'\s*\(\d+\s*л\.с\.\)', '', regex=True)
```

```python
rename_dict = {
    'Brand': 'brand',
    'Model': 'model',
    'Price': 'price',
    'Address': 'address',
    'Год выпуска': 'year',
    'Пробег': 'mileage (km)',
    'Владельцев по ПТС': 'num_owners',
    'Состояние': 'condition',
    'Модификация': 'modification',
    'Объём двигателя': 'engine_volume',
    'Тип двигателя': 'engine_type',
    'Коробка передач': 'transmission',
    'Привод': 'drive_type',
    'Комплектация': 'equipment',
    'Тип кузова': 'body_type',
    'Цвет': 'color'
}

df = df.rename(columns=rename_dict)
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# Получаем курсы валют с открытого API Центрального Банка России
# API возвращает JSON с актуальными курсами валют относительно рубля
url = "https://www.cbr-xml-daily.ru/daily_json.js"
response = requests.get(url)
data = response.json()

usd_rate = data["Valute"]["USD"]["Value"]
eur_rate = data["Valute"]["EUR"]["Value"]

df["price_USD"] = df["price"] / usd_rate
df["price_EUR"] = df["price"] / eur_rate
```

```python
# Обогатим данные про бренды автомобилей
unique_brands = df["brand"].unique()
unique_brands_df = pd.DataFrame(unique_brands, columns=["brand"])

unique_brands_df
```

```python
# Будем использовать американский API:
# NHTSA - National Highway Traffic Safety Administration

logging.getLogger().handlers.clear()

# Настройка логирования
logging.basicConfig(
    format="%(asctime)s - %(levelname)s - %(message)s",
    level=logging.INFO,
    stream=sys.stdout
)


# Функция для стандартизации названий брендов
def standardize_brand(brand):
    logging.info(f"Стандартизация бренда {brand}")
    url = "https://vpic.nhtsa.dot.gov/api/vehicles/GetAllMakes?format=json"
    try:
        response = requests.get(url).json()
        sleep(0.5)
        official_names = {m['Make_Name'].lower(): m['Make_Name']
                          for m in response['Results']}
        standardized = official_names.get(brand.lower(), brand)
        logging.info(f"Бренд {brand} стандартизирован в {standardized}")
        return standardized
    except Exception as e:
        logging.error(f"Ошибка при стандартизации бренда {brand}: {e}")
        return brand

# Функция для получения типов ТС, выпускаемых брендом
def get_vehicle_type(row):
    brand = row['brand']
    logging.info(f"Получение типа транспортного средства для бренда {brand}")
    url = f"https://vpic.nhtsa.dot.gov/api/vehicles/GetVehicleTypesForMake/{brand}?format=json"
    try:
        response = requests.get(url).json()
        sleep(0.5)
        vehicle_types = ', '.join(set([t['VehicleTypeName'] for t in response['Results']]))
        logging.info(f"Для бренда {brand} найден(ы): {vehicle_types}")
        return vehicle_types
    except Exception as e:
        logging.error(f"Ошибка при получении типа для бренда {brand}: {e}")
        return 'Unknown'

tqdm.pandas(desc="Processing")

unique_brands_df['brand_standardized'] = unique_brands_df['brand'].progress_apply(standardize_brand)
unique_brands_df['vehicle_types'] = unique_brands_df.progress_apply(get_vehicle_type, axis=1)

logging.info("Обработка завершена. Итоговый датафрейм:")
```

```python
# Обработаем колонку с типами автомобилей
unique_brands_df['vehicle_types'] = unique_brands_df['vehicle_types'].fillna("")

# Разбиваем строки по запятой и используем one hot encoding
vehicle_dummies = unique_brands_df['vehicle_types'].str.get_dummies(sep=", ")

unique_brands_encoded = pd.concat([unique_brands_df.drop(columns=["vehicle_types"]), vehicle_dummies], axis=1)
```

```python
import time

# Функция для геокодирования адреса с использованием API Nominatim (OpenStreetMap)
def geocode(address):
    base_url = "https://nominatim.openstreetmap.org/search"
    params = {
        "format": "json",
        "q": address,
        "addressdetails": 1,
        "limit": 1
    }
    headers = {"User-Agent": "Mozilla/5.0 (compatible; DataEnrichmentScript)"}
    try:
        logging.info(f"Отправка запроса к API для адреса {address}")
        response = requests.get(base_url, params=params, headers=headers, timeout=10)
        if response.status_code == 200:
            results = response.json()
            if results:
                lat = results[0].get("lat")
                lon = results[0].get("lon")
                logging.info(f"Успешно получены координаты для адреса {address}: широта {lat}, долгота {lon}")
                return lat, lon
            else:
                logging.warning(f"Нет результатов для адреса {address}")
        else:
            logging.error(f"Запрос для адреса {address} завершился со статус-кодом {response.status_code}")
    except Exception as e:
        logging.error(f"Ошибка при запросе для адреса {address}: {e}")
    return None, None

# Применяем геокодирование для каждого адреса и добавляем столбцы latitude и longitude
latitude_list = []
longitude_list = []
total = len(df["address"])
for i, addr in enumerate(df["address"], start=1):
    logging.info(f"Обработка адреса {i} из {total}: {addr}")
    lat, lon = geocode(addr)
    latitude_list.append(lat)
    longitude_list.append(lon)
    # Чтобы не перегружать API, делаем небольшую задержку
    time.sleep(1)

df["latitude"] = latitude_list
df["longitude"] = longitude_list

logging.info("Геокодирование завершено")
df.head()
```
